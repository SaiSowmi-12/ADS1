````
# ============================
# WHO + World Bank MERGE + EDA
# ============================
# Files expected in the working folder:
#   - "world sucide.csv"   (WHO suicide dataset; note: 'sucide' typo handled)
#   - "world bank.csv"     (World Bank indicators: Country Name, Series Name, year columns like '1980 [YR1980]')
#
# Output:
#   - merged_who_worldbank.csv
#   - basic plots (shown inline if in Jupyter)
# ============================

import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ----------------------------
# 0) Helper functions
# ----------------------------
def normalize_country(s: str) -> str:
    if pd.isna(s):
        return s
    s = str(s).strip()
    # unify punctuation/case
    s = re.sub(r"\s+", " ", s)
    # common harmonizations (expand as needed)
    fixes = {
        "United States": "United States of America",
        "USA": "United States of America",
        "US": "United States of America",
        "Russian Federation": "Russia",
        "Korea, Rep.": "South Korea",
        "Korea, Dem. People's Rep.": "North Korea",
        "Egypt, Arab Rep.": "Egypt",
        "Iran, Islamic Rep.": "Iran",
        "Venezuela, RB": "Venezuela",
        "Hong Kong SAR, China": "Hong Kong",
        "Gambia, The": "Gambia",
        "Congo, Dem. Rep.": "Democratic Republic of the Congo",
        "Congo, Rep.": "Republic of the Congo",
        "Yemen, Rep.": "Yemen",
        "Syrian Arab Republic": "Syria",
        "Bahamas, The": "Bahamas",
        "Brunei Darussalam": "Brunei",
        "Cote d'Ivoire": "Cte dIvoire",
        "Cte d'Ivoire": "Cte dIvoire",
        "Czech Republic": "Czechia",
        "Lao PDR": "Laos",
        "Slovak Republic": "Slovakia",
        "Eswatini": "Swaziland",
        "Macedonia, FYR": "North Macedonia"
    }
    return fixes.get(s, s)

def extract_year(colname: str):
    # Accept "1980 [YR1980]"  1980, or "1995"  1995
    m = re.search(r"(\d{4})", str(colname))
    return int(m.group(1)) if m else None

# ----------------------------
# 1) Load WHO suicide dataset
# ----------------------------
who_path_try = ["world sucide.csv", "world suicide.csv"]  # handle both spellings
who_df = None
for p in who_path_try:
    try:
        who_df = pd.read_csv(p)
        print(f"Loaded WHO file: {p}")
        break
    except Exception as e:
        pass
if who_df is None:
    raise FileNotFoundError("WHO file not found. Place 'world sucide.csv' (or 'world suicide.csv') next to this script.")

# standardize cols
who_df.columns = [c.strip().lower().replace(" ", "_") for c in who_df.columns]

# Expect these columns from your screenshot:
# country, year, sex, age, suicides_n, population
needed = {"country", "year", "suicides_n", "population"}
missing = needed - set(who_df.columns)
if missing:
    raise ValueError(f"WHO file missing expected columns: {missing}. Found: {who_df.columns.tolist()}")

# Clean basic types
who_df["country"] = who_df["country"].astype(str).map(normalize_country)
who_df["year"] = pd.to_numeric(who_df["year"], errors="coerce").astype("Int64")
who_df["suicides_n"] = pd.to_numeric(who_df["suicides_n"], errors="coerce")
who_df["population"] = pd.to_numeric(who_df["population"], errors="coerce")

# Aggregate to country-year (sum over sex/age groups), then compute per 100k rate
who_cy = (
    who_df.groupby(["country", "year"], dropna=True)[["suicides_n", "population"]]
    .sum()
    .reset_index()
)
who_cy = who_cy[who_cy["population"] > 0]
who_cy["suicides_per_100k"] = who_cy["suicides_n"] / who_cy["population"] * 100_000

print("WHO (country-year) shape:", who_cy.shape)
print(who_cy.head())

# ----------------------------
# 2) Load & reshape World Bank indicators
# ----------------------------
wb = pd.read_csv("world bank.csv")
# Keep only the columns we need (robust to column name variants)
cols = wb.columns.tolist()
# Identify year-like columns (those that have a 4-digit year somewhere, e.g., '1980 [YR1980]')
year_cols = [c for c in cols if extract_year(c) is not None]
id_candidates = ["Country Name", "Country", "country", "Series Name", "Series Name ", "Series Code", "Series Code "]
id_vars = [c for c in id_candidates if c in cols]

if not {"Country Name", "Series Name"}.issubset(set(id_vars)):
    print("Heads up: couldn't find expected 'Country Name'/'Series Name' exactly; using closest matches:", id_vars)

# Melt: wide years  long rows
wb_long = wb.melt(
    id_vars=id_vars,
    value_vars=year_cols,
    var_name="year_raw",
    value_name="value"
)
wb_long["year"] = wb_long["year_raw"].apply(extract_year).astype("Int64")
# Keep only real years
wb_long = wb_long.dropna(subset=["year"]).drop(columns=["year_raw"])

# Choose which field to pivot on for clean column names:
# Prefer 'Series Code' if present; otherwise use 'Series Name'
pivot_key = "Series Code" if "Series Code" in wb_long.columns else "Series Name"

wb_pivot = (
    wb_long.pivot_table(
        index=["Country Name", "year"],
        columns=pivot_key,
        values="value",
        aggfunc="first"
    )
    .reset_index()
)
wb_pivot.columns.name = None

# Normalize country + year
wb_pivot.rename(columns={"Country Name": "country"}, inplace=True)
wb_pivot["country"] = wb_pivot["country"].astype(str).map(normalize_country)
wb_pivot["year"] = pd.to_numeric(wb_pivot["year"], errors="coerce").astype("Int64")

# Optional: keep a core subset of well-known codes if available
preferred_codes = [
    "NY.GDP.PCAP.CD",     # GDP per capita (current US$)
    "SL.UEM.TOTL.ZS",     # Unemployment (% of total labor force)
    "SH.XPD.CHEX.GD.ZS",  # Current health expenditure (% of GDP)
    "SP.DYN.LE00.IN",     # Life expectancy at birth, total (years)
    "EN.POP.DNST",        # Population density (people per sq. km of land area)
    "SP.POP.65UP.TO.ZS"   # Population ages 65 and above (% of total)
]
available_pref = [c for c in preferred_codes if c in wb_pivot.columns]
wb_model = wb_pivot[["country", "year"] + (available_pref if available_pref else [c for c in wb_pivot.columns if c not in ["country","year"]])]

print("World Bank (pivot) shape:", wb_model.shape)
print(wb_model.head())

# ----------------------------
# 3) Merge WHO + World Bank
# ----------------------------
merged = pd.merge(
    who_cy,
    wb_model,
    on=["country", "year"],
    how="inner"
)

# Basic cleaning after merge
# Remove rows with too many missing features (allow a few NaNs)
feature_cols = [c for c in merged.columns if c not in ["country", "year", "suicides_n", "population", "suicides_per_100k"]]
min_non_na = max(1, int(0.6 * len(feature_cols)))  # keep rows with >=60% non-null features
keep_mask = merged[feature_cols].notna().sum(axis=1) >= min_non_na
merged = merged[keep_mask].copy()

print("\nMerged shape:", merged.shape)
print(merged.head())

# Save for modeling
merged.to_csv("merged_who_worldbank.csv", index=False)
print("\nSaved merged dataset  merged_who_worldbank.csv")

# ----------------------------
# 4) Quick EDA
# ----------------------------
# Global trend of suicide rate over time
plt.figure(figsize=(7,4))
merged.groupby("year")["suicides_per_100k"].mean().plot(marker="o")
plt.title("Global Mean Suicide Rate Over Time")
plt.xlabel("Year"); plt.ylabel("Suicides per 100k (mean)")
plt.tight_layout(); plt.show()

# Top 10 countries by average suicide rate across years
top10 = (
    merged.groupby("country")["suicides_per_100k"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
)
plt.figure(figsize=(8,5))
top10.sort_values().plot(kind="barh")
plt.title("Top 10 Countries by Avg Suicide Rate (per 100k)")
plt.xlabel("Avg suicides per 100k"); plt.ylabel("")
plt.tight_layout(); plt.show()

# Correlation (numeric features only)
num_cols = merged.select_dtypes(include=[np.number]).columns
corr = merged[num_cols].corr()

plt.figure(figsize=(9,7))
plt.imshow(corr, aspect="auto")
plt.title("Correlation Heatmap (numeric features)")
plt.colorbar(label="Correlation")
plt.xticks(range(len(num_cols)), num_cols, rotation=90)
plt.yticks(range(len(num_cols)), num_cols)
plt.tight_layout(); plt.show()

# Print missingness summary
missing_summary = merged.isna().mean().sort_values(ascending=False).head(20)
print("\nTop columns by missingness (proportion of NaN):")
print(missing_summary)
````

````
# ============================
# WHO + WORLD BANK MERGE + EDA
# ============================

import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ---------- Helpers ----------
def normalize_country(name: str) -> str:
    """Normalize country names to improve joins."""
    if pd.isna(name):
        return name
    s = str(name).strip()
    s = re.sub(r"\s+", " ", s)
    fixes = {
        "United States": "United States of America",
        "USA": "United States of America",
        "US": "United States of America",
        "Russian Federation": "Russia",
        "Korea, Rep.": "South Korea",
        "Korea, Dem. People's Rep.": "North Korea",
        "Egypt, Arab Rep.": "Egypt",
        "Iran, Islamic Rep.": "Iran",
        "Venezuela, RB": "Venezuela",
        "Hong Kong SAR, China": "Hong Kong",
        "Gambia, The": "Gambia",
        "Congo, Dem. Rep.": "Democratic Republic of the Congo",
        "Congo, Rep.": "Republic of the Congo",
        "Syrian Arab Republic": "Syria",
        "Bahamas, The": "Bahamas",
        "Brunei Darussalam": "Brunei",
        "Cote d'Ivoire": "Cte dIvoire",
        "Cte d'Ivoire": "Cte dIvoire",
        "Czech Republic": "Czechia",
        "Lao PDR": "Laos",
        "Slovak Republic": "Slovakia",
        "Eswatini": "Swaziland",
        "Macedonia, FYR": "North Macedonia"
    }
    return fixes.get(s, s)

def extract_year(colname: str):
    """Extract a 4-digit year from a column header like '1980 [YR1980]'."""
    m = re.search(r"(\d{4})", str(colname))
    return int(m.group(1)) if m else None
````

````

# ============================
# Handle both spellings just in case
for path in ["world sucide.csv", "world suicide.csv"]:
    try:
        who_df = pd.read_csv(path)
        print(f"Loaded WHO file: {path}")
        break
    except Exception:
        who_df = None
if who_df is None:
    raise FileNotFoundError("Place 'world sucide.csv' (or 'world suicide.csv') next to this notebook.")

# standardize column names
who_df.columns = [c.strip().lower().replace(" ", "_") for c in who_df.columns]

# Patch: accept suicides_no (your file) or suicides_n
if "suicides_n" not in who_df.columns and "suicides_no" in who_df.columns:
    who_df.rename(columns={"suicides_no": "suicides_n"}, inplace=True)

# Patch: accept Country  country
if "country" not in who_df.columns and "country_name" in who_df.columns:
    who_df.rename(columns={"country_name": "country"}, inplace=True)

# Validate required columns
needed = {"country", "year", "suicides_n", "population"}
missing = needed - set(who_df.columns)
if missing:
    raise ValueError(f"WHO file missing expected columns: {missing}. Columns found: {who_df.columns.tolist()}")

# Types & cleaning
who_df["country"]     = who_df["country"].astype(str).map(normalize_country)
who_df["year"]        = pd.to_numeric(who_df["year"], errors="coerce").astype("Int64")
who_df["suicides_n"]  = pd.to_numeric(who_df["suicides_n"], errors="coerce")
who_df["population"]  = pd.to_numeric(who_df["population"], errors="coerce")

# Aggregate to country-year (sum across sex/age), compute rate
who_cy = (
    who_df.groupby(["country", "year"], dropna=True)[["suicides_n", "population"]]
    .sum()
    .reset_index()
)
who_cy = who_cy[who_cy["population"] > 0].copy()
who_cy["suicides_per_100k"] = who_cy["suicides_n"] / who_cy["population"] * 100_000

print("WHO (country-year) shape:", who_cy.shape)
display(who_cy.head())

````

````
# ==================================
# 2) Load & reshape World Bank (WDI)
# ==================================
import numpy as np
import pandas as pd
import re

# Treat ".." as missing
wb = pd.read_csv("world bank.csv", keep_default_na=True, na_values=[".."])

# 1) Strip whitespace from ALL column headers (fixes 'Country '  'Country')
wb.columns = [c.strip() for c in wb.columns]

# 2) Accept either 'Country Name' or 'Country'
country_col = "Country Name" if "Country Name" in wb.columns else (
    "Country" if "Country" in wb.columns else None
)
if country_col is None:
    raise ValueError(f"Could not find a country column. Columns: {wb.columns.tolist()}")

# 3) Accept either 'Series Code' (preferred) or 'Series Name'
series_code_col = "Series Code" if "Series Code" in wb.columns else None
series_name_col = "Series Name" if "Series Name" in wb.columns else None
if series_code_col is None and series_name_col is None:
    raise ValueError("World Bank file must contain either 'Series Code' or 'Series Name'.")

# 4) Build id_vars dynamically (include Country Code if present; harmless extra id)
id_vars = [country_col]
if "Country Code" in wb.columns: id_vars.append("Country Code")
if series_code_col: id_vars.append(series_code_col)
if series_name_col: id_vars.append(series_name_col)

# 5) Identify year-like columns (e.g., '1980 [YR1980]', '2022 [YR2022]')
def extract_year(colname: str):
    m = re.search(r"(\d{4})", str(colname))
    return int(m.group(1)) if m else None

year_cols = [c for c in wb.columns if extract_year(c) is not None and c not in id_vars]
if not year_cols:
    raise ValueError("No year columns found (like '1980 [YR1980]').")

# 6) Melt widelong
wb_long = wb.melt(
    id_vars=id_vars,
    value_vars=year_cols,
    var_name="year_raw",
    value_name="value"
)
wb_long["year"] = wb_long["year_raw"].apply(extract_year).astype("Int64")
wb_long = wb_long.dropna(subset=["year"]).drop(columns=["year_raw"])

# 7) Choose series label for pivot (codes if available, else names)
pivot_key = series_code_col if series_code_col is not None else series_name_col

# 8) Pivot so each indicator becomes its own column
wb_pivot = (
    wb_long.pivot_table(
        index=[country_col, "year"],
        columns=pivot_key,
        values="value",
        aggfunc="first"
    )
    .reset_index()
)
wb_pivot.columns.name = None

# 9) Standardize for merge
wb_pivot.rename(columns={country_col: "country"}, inplace=True)
wb_pivot["country"] = wb_pivot["country"].astype(str)
wb_pivot["year"]    = wb_pivot["year"].astype("Int64")

# 10) Keep a useful subset if available (fallback to all columns)
preferred_codes = [
    "NY.GDP.PCAP.CD",     # GDP per capita (current US$)
    "SL.UEM.TOTL.ZS",     # Unemployment (% of total labor force)
    "SH.XPD.CHEX.GD.ZS",  # Health expenditure (% of GDP)
    "SP.DYN.LE00.IN",     # Life expectancy at birth
    "EN.POP.DNST",        # Population density
    "SP.POP.65UP.TO.ZS"   # % population aged 65+
]
available = [c for c in preferred_codes if c in wb_pivot.columns]
wb_model = wb_pivot[["country", "year"] + (available if available else [c for c in wb_pivot.columns if c not in ["country","year"]])]

print("World Bank (pivot) shape:", wb_model.shape)
display(wb_model.head())
````

````
#  Merge WHO and World Bank datasets
merged = pd.merge(who_cy, wb_model, on=["country", "year"], how="inner")

print("Merged dataset shape:", merged.shape)
display(merged.head())

# Optional: drop rows with NaN in target or key feature
merged = merged.dropna(subset=["suicides_per_100k", "SH.XPD.CHEX.GD.ZS"])

# Save cleaned merged data
merged.to_csv("merged_final.csv", index=False)
print(" Saved merged dataset as merged_final.csv")
````

````
def recursive_forecast(df_hist, model, feat_cols, target="suicides_per_100k", steps=5):
    """
    df_hist: historical rows including the latest year per country (with lag features already computed)
    model: fitted model
    feat_cols: columns used by the model (must include lag features)
    steps: number of future years to predict
    """
    out = []
    latest_year = int(df_hist["year"].max())

    # we keep only the last year per country as the rolling base
    base = df_hist[df_hist["year"] == latest_year].copy()

    # work copy with all rows well append predictions to (to recompute lags each step)
    work = df_hist.copy()

    for h in range(1, steps + 1):
        # build the feature frame for countries that have complete lag info at this point
        X_next = base[feat_cols].copy().dropna()
        if X_next.empty:
            break

        preds = model.predict(X_next)
        tmp = base.loc[X_next.index, ["country","year"]].copy()
        tmp["year"] = latest_year + h
        tmp[target] = preds

        # append to work so that lags can reference these newly predicted values
        work = pd.concat([work, tmp], ignore_index=True, sort=False)

        # re-compute the lag features using the same routine you used originally
        work = work.sort_values(["country","year"]).copy()
        for c in [target] + [c for c in feat_cols if c.endswith("_lag1") is False]:
            # only re-lag the target (and any non-lag base drivers if you used them directly)
            # simplest: re-create only target lag1
            pass
        # Minimal: ensure target lag1 is available
        work["suicides_per_100k_lag1"] = work.groupby("country")[target].shift(1)

        # next base = current steps year rows
        base = work[work["year"] == latest_year + h].copy()

        tmp["forecast_horizon"] = h
        out.append(tmp)

    return pd.concat(out, ignore_index=True) if out else pd.DataFrame()

# Example use after fitting your selected/best pipeline
best_pipe.fit(X_train, y_train)
hist = pd.concat([train, test], axis=0)  # or just train if you want strictly pre-test history
multi_fcst = recursive_forecast(hist, best_pipe, X_cols, steps=5)

# optional: pick 35 years as required
fcst_3_to_5 = multi_fcst[multi_fcst["forecast_horizon"].between(3,5)]
````

````
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
import pandas as pd

def evaluate(name, y_true, y_pred):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae  = mean_absolute_error(y_true, y_pred)
    r2   = r2_score(y_true, y_pred)
    return {"Model": name, "RMSE": rmse, "MAE": mae, "R2": r2}

# after predicting on X_test:
pred_lr   = lr_pipe.predict(X_test)
pred_rf   = rf_pipe.predict(X_test)
pred_lgb  = lgb_pipe.predict(X_test)
pred_cat  = cat_pipe.predict(X_test)

res = pd.DataFrame([
    evaluate("LinearRegression", y_test, pred_lr),
    evaluate("RandomForest",    y_test, pred_rf),
    evaluate("LightGBM",        y_test, pred_lgb),
    evaluate("CatBoost",        y_test, pred_cat),
]).sort_values("RMSE")
print(res)
````
